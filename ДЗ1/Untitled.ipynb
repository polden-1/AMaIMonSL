```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import Eq, symbols, Matrix, eye, latex, linsolve
from sympy import solve as sysolve
from scipy.linalg import solve
from IPython.display import Math
```

```python
matrix = np.loadtxt("data.csv", delimiter="\t")
display(Math(r"\text{Матрица переходов}"), Matrix(matrix))
```

```python
node_cnt = len(matrix)
pt_e = (matrix.T - eye(node_cnt))
for i in range(node_cnt):
    pt_e[-1, i] = 1.

b_e = Matrix(np.zeros((node_cnt, 1)))
b_e[-1] = 1.

vect_lim = list(linsolve((pt_e, b_e)))[0]

display(Math(r"\text{Вектор предельных вероятностей   }"), Matrix(vect_lim))
```

```python
matrix_lim = Matrix([vect_lim for i in range(node_cnt)])
display(Math(r"\text{Предельная матрица прееходов   }"), matrix_lim)
```

```python
imitation_cnt = 50
shift_cnt = 100
position_entry_matrix = []

program_entry_matrix = np.zeros((node_cnt, node_cnt))
for i in range(node_cnt):
    program_entry_matrix[i][0] = matrix[i][0]
    for j in range(1, node_cnt):
        program_entry_matrix[i][j] = program_entry_matrix[i][j-1] + matrix[i][j]

for i in range(imitation_cnt):
    start_position = np.random.default_rng(i).integers(0,4)
    display(Math(r"\text{Стартовая позиция для начала моделирования} = " + str(start_position + 1)))
    
    current_position = start_position
    prev_position = -1
    
    position_history = [[0, int(start_position)]]
    position_entry = {}
    
    for j in range(shift_cnt):
        current_prob = np.random.default_rng(imitation_cnt + j).integers(1,100)/100
        next_position = 0
        while current_prob > program_entry_matrix[current_position][next_position]:
            next_position += 1
        
        prev_position = current_position
        current_position = next_position
        position_entry[current_position] = position_entry.get(current_position, 0) + 1

        position_history.append([j + 1, current_position + 1])
    
    steps, positions = zip(*position_history)
    plt.plot(steps, positions, marker='o', linestyle='-', color='b', label="Позиция во времени")
    plt.xlabel("Шаги")
    plt.ylabel("Позиция")
    plt.title("График движения")
    plt.legend()
    plt.grid(True)
    plt.show()

    latex_table = r"\begin{array}{|c|c|} \hline"
    latex_table += r"\text{Позиция} & \text{Количество вхождений} \\ \hline"
    position_entry = dict(sorted(position_entry.items()))
    for pos, count in position_entry.items():
        latex_table += f"{pos+1} & {count} \\\\ \\hline"
    latex_table += r"\end{array}"
    display(Math(latex_table))

    position_entry_matrix.append([[i], [position_entry[k] for k in range(node_cnt)]])

    
```

```python
steps, positions = zip(*position_entry_matrix)
plt.plot(steps, positions)
plt.grid(True)
plt.show()
```
